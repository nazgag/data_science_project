```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
population_data = pd.read_csv('/home/mac/Downloads/population2010-2022byregionnationalityandgendercsv-1.csv')

# Group data by year and calculate total population
population_evolution = population_data.groupby('Year')['Population'].sum()

# Plot population evolution
plt.figure(figsize=(10, 6))
population_evolution.plot(kind='line', marker='o', color='b')
plt.title('Population Evolution in Saudi Arabia (2010-2022)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```

```python
Question 2: Gender Distribution
Next, we explore the gender distribution across different regions. A bar chart will display the proportion of males and females in each region.
```

```python
# Code for gender distribution analysis
gender_distribution = population_data.groupby(['Region', 'Gender'])['Population'].sum().unstack()

# Plot gender distribution
gender_distribution.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Gender Distribution Across Regions')
plt.xlabel('Region')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.show()
```
